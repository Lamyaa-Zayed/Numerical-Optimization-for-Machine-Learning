##### Input data (X), Target Labels (y), Learning Rate (alpha), Maximum number of iterations

```python
theta_0 = 0
theta_1 = 1
alpha = 0.005   #learning rate
max_iteration = 5000
```

```python
def SVLR (x, y, alpha, max_iteration):
    theta_0 = 0
    theta_1 = 1

    cost_function = []
    theta0_values = []
    theta1_values = []
    new_hypothesis = []
    hypothesis = np.zeros(len(x))
    
    for i in range (max_iteration):
        
        hypothesis = theta_0 + theta_1 * x
        new_hypothesis.append(hypothesis)
        
        cost = (1/2*len(x))* np.sum(((hypothesis - y) ** 2))
        cost_function.append(cost)
        print(cost)
        
        gradiant_theta0 = (1/len(x)) * np.sum(hypothesis - y)
        gradiant_theta1 = (1/len(x)) * np.sum((hypothesis - y) * x)
        
        theta_0 = theta_0 - alpha * gradiant_theta0
        theta0_values.append(theta_0)
        
        theta_1 = theta_1 - alpha * gradiant_theta1
        theta1_values.append(theta_1)
                             
        #gradiant_norm = np.sqrt((gradiant_theta0 ** 2)+(gradiant_theta1 ** 2))
        #if(gradiant_norm == 0):
        #    break
        if(cost < 0.05):
            final_epoch = i
            print("final epoch= ",final_epoch)
            break
        
    
    return theta0_values, theta1_values, cost_function, new_hypothesis
```

#### Update the implementation of the Batch GD to be Mini-Batch and rpeat the above plots and evaluations
#### Change batch size and see the effect

```python
def mini_batch (x, y, alpha, epoch, batch_size):
    theta_0 = 0
    theta_1 = 1

    cost_function = []
    theta0_values = []
    theta1_values = []
    new_hypothesis = []
    hypothesis = np.zeros(len(x))
    num_of_batches = int(len(x) / batch_size)
    
    for i in range (epoch):
        for j in range(0, num_of_batches):  
            
            hypothesis = theta_0 + theta_1 * x[j*batch_size : (j+1)*batch_size]
            new_hypothesis.append(hypothesis)

            cost = (1/2*len(x))* np.sum(((hypothesis - y[j*batch_size : (j+1)*batch_size]) ** 2))
            cost_function.append(cost)
            print(cost)

            gradiant_theta0 = (1/len(x)) * np.sum(hypothesis - y[j*batch_size : (j+1)*batch_size])
            gradiant_theta1 = (1/len(x)) * np.sum((hypothesis - y[j*batch_size : (j+1)*batch_size]) * x[j*batch_size : (j+1)*batch_size])

            theta_0 = theta_0 - alpha * gradiant_theta0
            theta0_values.append(theta_0)

            theta_1 = theta_1 - alpha * gradiant_theta1
            theta1_values.append(theta_1)

            #gradiant_norm = np.sqrt((gradiant_theta0 ** 2)+(gradiant_theta1 ** 2))
            #if(gradiant_norm == 0):
            #    break
            if(cost < 0.05):
                final_epoch = i
                print("final epoch= ",final_epoch)
                break


    return theta0_values, theta1_values, cost_function, new_hypothesis
```

```python
def stochastic (x, y, alpha, epoch):
    theta_0 = 0
    theta_1 = 1

    cost_function = []
    theta0_values = []
    theta1_values = []
    new_hypothesis = []
    hyp = []
    #num_of_batches = int(len(x) / batch_size)
    
    for i in range (epoch):
        for j in range(len(x)): 
            
            hypothesis = theta_0 + theta_1 * x[j]
            new_hypothesis.append(hypothesis)

            cost = ((hypothesis - y[j]) ** 2)
            cost_function.append(cost)
            print(cost)

            gradiant_theta0 = 2 * (hypothesis - y[j])
            gradiant_theta1 = 2 * ((hypothesis - y[j]) * x[j])

            theta_0 = theta_0 - alpha * gradiant_theta0
            theta0_values.append(theta_0)

            theta_1 = theta_1 - alpha * gradiant_theta1
            theta1_values.append(theta_1)

            #gradiant_norm = np.sqrt((gradiant_theta0 ** 2)+(gradiant_theta1 ** 2))
            #if(gradiant_norm == 0):
            #    break
            if(cost < 0.05):
                final_epoch = i
                print("final epoch= ",final_epoch)
                break

        
    return theta0_values, theta1_values, cost_function, new_hypothesis
```
